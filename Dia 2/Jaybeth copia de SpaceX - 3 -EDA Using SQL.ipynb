### Ejercicio 2

##### Muestra 5 lanzamientos cuya plataforma empiece por 'CCS'


```python
queries = "SELECT LaunchSite FROM `SPACEXTBL` WHERE LaunchSite LIKE ('CCS%')"
df  = pd.read_sql(queries, con = engine )
df
```

### Ejercicio 3

##### Mustra el la masa total transportada por los lanzamientos realizados donde el cliente es NASA (CRS)


```python
queries = "SELECT SUM(`PayloadMass`) FROM `SPACEXTBL` WHERE `Customers` = 'NASA (CRS)'"
df  = pd.read_sql(queries, con = engine )
df
```

### Ejercicio 4

##### Muestra la masa media tranportada por lanzamientos con serie que comienza por B1

```python
queries = "SELECT AVG(`PayloadMass`) FROM `SPACEXTBL` WHERE Serial LIKE ('B1%')"
df  = pd.read_sql(queries, con = engine )
df
```

### Ejercicio 5

##### Muestra la fecha donde se realizo el primer aterrizaje exitoso sobre terreno (TRUE RTLS)

*Pista:Utiliza la función min*


```python
queries = "SELECT min(Date) FROM `SPACEXTBL` WHERE LandingOutcome = ('True RTLS')"
df  = pd.read_sql(queries, con = engine )
df
```

### Ejercicio 6

##### Muestra las series de los cohetes que aterrizaron sobre portaviones (TRUE ASDS) con una masa entre 4000 y 6000

```python
queries = "SELECT SERIAL FROM `SPACEXTBL` WHERE LandingOutcome = ('True ASDS') BETWEEN 4000 AND 6000"
df  = pd.read_sql(queries, con = engine )
df
```

### Ejercicio 7

##### Muestra el numero de lanzamientos que completaron su mision correctamente y aquellos que fallaron

```python
queries = "SELECT COUNT(Mission_Outcome), Mission_Outcome FROM SPACEXTBL GROUP BY Mission_Outcome"

df  = pd.read_sql(queries, con = engine )
df
```

### Ejercicio 8

##### Muestra las series de los cohetes que transportaron la masa máxima

```python
queries = "SELECT MAX(PayloadMass) FROM SPACEXTBL"

df  = pd.read_sql(queries, con = engine )
df
```

```python
queries = "SELECT SERIAL FROM SPACEXTBL WHERE PayloadMass = (SELECT MAX(PayloadMass) FROM SPACEXTBL)"

df  = pd.read_sql(queries, con = engine )
df
```

```python
queries = "SELECT SERIAL, MAX(PayloadMass)  FROM SPACEXTBL GROUP BY SERIAL ORDER BY 2"

df  = pd.read_sql(queries, con = engine )
df
```
